```python
wc_summ = pd.read_csv("/content/sample_data/FIFA - World Cup Summary.csv", index_col="YEAR")
wc_summ.head(25)
```

```python
# Which team has won the most World Cups?
wins_by_team = wc_summ.groupby("CHAMPION").size().reset_index(name="Wins")
```

```python
wins_by_team = wins_by_team.sort_values("Wins", ascending=False)
print(wins_by_team.head(10))
```

```python
sns.barplot(x="CHAMPION", y="Wins", data=wins_by_team)
plt.xlabel("Team")
plt.ylabel("Number of World Cup Wins")
plt.title("Number of World Cup Wins by Team")
plt.xticks(rotation=45)
plt.show()
```

```python
# Does being host give a better chance of a World Cup Win
host_won = []
for index, row in wc_summ.iterrows():
  host_won.append(row["HOST"] == row["CHAMPION"])
wc_summ["HOST WON"] = host_won
```

```python
sns.countplot(x="HOST WON", data=wc_summ)
plt.xlabel("Host Won")
plt.ylabel("Number of Tournaments")
plt.title("Number of World Cup Tournaments Won by Host")
plt.show()
```

```python
# Count the number of times the host team won the World Cup
host_wins = wc_summ["HOST WON"].sum()

# Count the total number of World Cup tournaments
total_tournaments = wc_summ.shape[0]

# Calculate the percentage of host country wins
percentage_host_wins = (host_wins / total_tournaments) * 100
print(f"Percentage of host country wins: {percentage_host_wins:.2f}%")
```

```python
# Find the unique teams that have joined the World Cup
df_list = [wc_1930, wc_1934, wc_1938, wc_1950, wc_1954, wc_1958, wc_1962, wc_1966, wc_1970, wc_1974, wc_1978, wc_1982, wc_1986, wc_1990, wc_1994, wc_1998, wc_2002, wc_2006,
wc_2010, wc_2014, wc_2018,wc_2022]
```

```python
list_of_teams = []
for tournament in df_list:
    for team in tournament["Team"]:
        list_of_teams.append(team)
unique_teams = list(set(list_of_teams))
print(f"There are {len(unique_teams)} unique teams out of the {len(list_of_teams)} that have participated in the history of the World Cup.")
```

```python
# Which team has the most World Cup appearances? Which ones have the least?
team_app={
    "Team":[],
    "Appearance":[]
}
for team in unique_teams:
  team_app["Team"].append(team)
  team_app["Appearance"].append(list_of_teams.count(team))
team_app_data=pd.DataFrame(team_app).sort_values(by="Appearance",ascending=False).reset_index(drop=True)
team_app_data.head()
```

```python
# How has goalscoring improved throughout the years?
goals_per_year = wc_summ.groupby("YEAR")["GOALS SCORED"].mean().reset_index(name="goals_per_match")

# Create a line plot of the goals per match by year
sns.lineplot(x="YEAR", y="goals_per_match", data=goals_per_year)

# Show the plot
plt.show()
```

```python
# How many goals do championship teams usually score and concede?
champion_data = {
    "Year": [1930,1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 1978, 1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014, 2018,2022],
    "Team": [],
    "Goals Scored": []
}

for tournament in df_list:
    champion_data["Team"].append(tournament.iloc[0]["Team"])
    champion_data["Goals Scored"].append(tournament.iloc[0]["Goals For"])
champion_df = pd.DataFrame(champion_data).set_index("Year")
champion_df.head()
```
